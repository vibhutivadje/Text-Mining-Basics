# 2. Chunking


Chunking is to specify structure using part of speech tagging and take hard or similar structures

```python
from nltk.tree import Tree
from IPython.display import display
```

```python
import nltk
text="Our team completed the project and the department received grant money"
sentences=nltk.sent_tokenize(text)
# print(sentences)
#some verbs following some nouns i.e noun and their action
for i in sentences:
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    chunkstyle="Chunk:{<NN.?><VB.?>}" #define the chunk style, <NN.> includes <NN>, <NNS>, and <NNP>, but not <NNPS>
#     chunkstyle="Chunk:{<NN.*><VB.?>+}" #NN,NNS,NNP,NNPS, VB,VBD,VBZ, VBG
#? - 0 or 1
#* - 0 or more , + - 1 or more

#<NN.?> NN NNS NNP doesn't include NNPS
#<NN.+> NNS NNP NNPS
#<NN.*> includes all the above
#+ means mandatory, ?* means optional
    chunkparser=nltk.RegexpParser(chunkstyle) #define the chunk parser
    chunked=chunkparser.parse(tagged)#list of tuple
    print(chunked)#in the output, the line starting with "Chunk" indicates the chunks found
#     chunked.draw() #visualize chunking result
#     display(chunked)
```

```python
# A is a B is a good way to learn the attributes of something eg A can be person or proper noun
#PRP include: he, she, I , it, we
import nltk
text2='''He is the President of the United States. Before entering politics,he was a businessman. 
'''
```

```python
sentences2=nltk.sent_tokenize(text2)
for i in sentences2:
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    chunkstyle="Chunk:{<PRP><VB.+><DT>?<NN.?>}" # <NN.?> include NN, but <NN.+> doesn't include NN, beware
    #<DT>? means DT(a) is optional, <JJ.?> matches all the adjectives JJ: JJ, JJS, JJR
    #chunkstyle="Chunk:{<PRP><VB.?>+<DT>?<JJ.?>*<NN.*>+}"
    chunkparser=nltk.RegexpParser(chunkstyle) 
    chunked=chunkparser.parse(tagged)
    print(chunked)
#     chunked.draw()
```

```python
# we can use adjective + noun to quickly learn about something
#Imp: useful for introductory paragraph
# let's take a look the combination of an adjective and a noun which is usually used
import nltk
text3='''By design, a blockchain is resistant to modification of the data.
It is "an open, distributed ledger that can record transactions
between two parties efficiently".[7]
For use as a distributed ledger, a blockchain is typically managed by a
peer-to-peer network collectively adhering to a protocol for inter-node
communication and validating new blocks. Once recorded, the data in any
given block cannot be altered retroactively without alteration of all
blocks, which requires consensus of the network majority.
Although blockchain records are not unalterable, blockchains is the key for 
secured transactions, and expedited varifications. However,
the Bitcoin payment network is
an unregulated network, which is always criticized for illegal transactions
'''
#<NN> - nothing after the bracket means only one
sentence=nltk.sent_tokenize(text3)
chunkstyle='Chunk:{<JJ.?><NN.*>}'
parser=nltk.RegexpParser(chunkstyle)
for i in sentence:
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    chunked=parser.parse(tagged)
    print(chunked)
#     chunked.draw()
```

```python
import nltk
text='''Donald John Trump (born June 14, 1946) is the 45th and
current President of the United States. Before entering politics,
he was a businessman and television personality. 
'''
chunkstyle="Chunk:{<PRP><VB.+><DT>?<NN.?>}" 
chunkparser=nltk.RegexpParser(chunkstyle) 
sentences=nltk.sent_tokenize(text)
for i in sentences:
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    chunked=chunkparser.parse(tagged)
    for subtree in chunked.subtrees():
#         print(subtree)
        if subtree.label()=='Chunk':
            print(subtree)
            print(subtree.leaves())
```

```python
# specify some word must be some specific words, A encompass B
import nltk
brown = nltk.corpus.brown

lst=['contain','contained','contains','cover','covers','covered','covering','hold','holds','held','carry','carries','carried']
import nltk
cp=nltk.RegexpParser("Chunk:{<NN.*><VB.+><DT>?<NN.*>+}")
brown = nltk.corpus.brown
for sent in brown.tagged_sents():
    tree=cp.parse(sent)
    for subtree in tree.subtrees():
        if subtree.label()=='Chunk' and subtree.leaves()[1][0] in lst:
            print(subtree.leaves())
```

```python
# another example, find the jjs in front of the word "people" in the inaugural speech
import nltk
from nltk.corpus import inaugural
text=inaugural.words()
pos=nltk.pos_tag(text)

#JJ+NN, NN will be people
jjlist=[]
cp=nltk.RegexpParser("Chunk:{<DT|JJ.?>+<NN.*>}") #couple of DT and JJ mixed together
brown = nltk.corpus.brown
for sent in brown.tagged_sents():
    tree=cp.parse(sent)
    for subtree in tree.subtrees():
        if subtree.label()=='Chunk' and subtree.leaves()[-1][0]=="people":
            jj=[z[0] for z in subtree.leaves()[:-1]]
            jjlist.extend(jj)
frequency=nltk.FreqDist(jjlist)
print(frequency.most_common(10)) # what kind of people are mostly talked about?
```

```python
# another example, find the verb phrase about the word "people" in the inaugural speech
#actions performed by people in speech
import nltk
from nltk.corpus import inaugural
text=inaugural.words()

#NN + VB phrase, NN=='people'
pos=nltk.pos_tag(text)
cp=nltk.RegexpParser("Chunk:{<NN.*><VB.+>+<DT|JJ.?|NN.*>+}")
brown = nltk.corpus.brown
for sent in brown.tagged_sents():
    tree=cp.parse(sent)
    for subtree in tree.subtrees():
        if subtree.label()=='Chunk' and subtree.leaves()[0][0]=="people":
            print(subtree.leaves())
```

# 4. Chinking

```python
# chinking is to exclude something from a chunk
# grab everything except
#chinking is to drop, delete the specified structure
import nltk
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),
       ("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
grammar = """
NP: {<.*>+}          # Chunk everything
}<VBD|IN>+{      # Chink sequences of VBD and IN
"""
cp = nltk.RegexpParser(grammar)
print(cp.parse(sentence))
cp.parse(sentence).draw()
print(cp.parse(sentence).label())
```

```python
import nltk
for i in nltk.sent_tokenize(text):
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    named_entity=nltk.ne_chunk(tagged)#recognises the name-entity eg:Person,Location,Date,Time
    print(named_entity)
#     named_entity.draw()
```

```python
def count_Nouns(text):
    c=0
    t=nltk.word_tokenize(text)
    p=nltk.pos_tag(t)
    for i in p:
        c+=i[1]in['NN','NNS','NNP','NNPS']
    return c
count_Nouns(text)
```

```python
# get a list of named entities

org=[]
person=[]
GPE=[]

import nltk
from nltk.tree import Tree
for i in nltk.sent_tokenize(text):
    words=nltk.word_tokenize(i)
    tagged=nltk.pos_tag(words)
    named_entity=nltk.ne_chunk(tagged)
   
    for i in named_entity.subtrees():
#         print(i)
        if i.label()=='ORGANIZATION':
            org.append(' '.join([w[0] for w in i.leaves()]))
        elif i.label()=='PERSON':
            person.append(' '.join([w[0] for w in i.leaves()]))
        elif i.label()=='GPE':
            GPE.append(' '.join([w[0] for w in i.leaves()]))
print(org)
print(person)
print(GPE)
```

# 11. find adjectives in front of the word "market"

```python
# there are a series of txt documents in the test folder, for each txt file, list all the adjectives in front of "market" to learn about the market without reading the articles.
#extract all adjectives in front of market.
import os
import re
import nltk
os.chdir('/Users/ashutoshfolane/Documents/VIBHUTI/CSUEB/BAN_675/test')#changes the directory
output=open('output.csv','w')
for i in os.listdir():#iterate through each file in that directory
    if i.endswith('.txt'):
        with open(i,'r') as f:#creating a file connection and read r mode
            contents=f.readlines()#creates a list of sentences 
            postcount=len(contents)-9#focus is on message which starts from 10th line
            adjectives=[]
            for j in range(9,len(contents)):
                line=contents[j]
                post=str(re.split('\t',line)[3])#text msg from the list
                words=nltk.word_tokenize(post)
                tagged=nltk.pos_tag(words)#tagged is a list of tuples
                ajective=[k[0] for k in tagged[:-1] if k[1][:2]=='JJ' and 'market' in tagged[tagged.index(k)+1][0]]
                adjectives+=ajective
            value=str(i)+'\t'+'\t'.join(adjectives)+'\n'
            print(value)
            output.write(value)
output.close()
```
